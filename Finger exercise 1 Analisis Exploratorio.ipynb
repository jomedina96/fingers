```python
#importo las bibliotecas 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#abro el archivo
datos=pd.read_csv("events.csv", low_memory=False)
```

```python
#obtengo una primera descripcion del archivo
datos.describe()
```

```python
#leo como son las columnas
datos.head(10)
```

```python
#problema 1
#valores distintos 
conteoDeEventos = datos["event"].value_counts().head(5)
```

```python
g = sns.barplot(x=conteoDeEventos.values, y=conteoDeEventos.index, orient='h')
g.set_title("Eventos predominantes", fontsize=15)
g.set_xlabel("Frecuencia", fontsize=12)
g.set_ylabel("Tipo de evento", fontsize=12)
```

```python
#problema2
d = datos.loc[datos['event'].str.contains("conversion")]
d = d.filter(["event","sku"])
conteoDeEventos = d["sku"].value_counts().head(5)
g = sns.barplot(x=conteoDeEventos.values, y=conteoDeEventos.index, orient='h',order=conteoDeEventos.index)
g.set_title("Productos con mayor cantidad de conversiones. ", fontsize=15)
g.set_xlabel("Cantidad de conversiones", fontsize=12)
g.set_ylabel("Productos(SKU)", fontsize=12)
```

```python
#problema 3
conteoDeEventos = datos["person"].value_counts().head(10)
g = sns.barplot(x=conteoDeEventos.values, y=conteoDeEventos.index, orient='h')
g.set_title("Usuarios con mayor actividad", fontsize=15)
g.set_xlabel("Cantidad de eventos", fontsize=12)
g.set_ylabel("Usuarios", fontsize=12)
```
